```python
#Matrix Multiplication
a=np.array([2,6,8,3,6]).reshape(1,5)
b=np.array([1,4,6,7,8]).reshape(5,1)
print(a)
print(b)
c=a@b
print(c)
```

```python
#Matrix multiplication
a=np.array([2,6,8,3,6]).reshape(1,5)
b=np.array([1,4,6,7,8]).reshape(5,1)
print(a)
print(b)
c=a.dot(b)
print(c)
```

```python
#max & min works only for numeric data types like integer and float
a=np.array([2,6,18,19,35])
print(a.max())
print(a.min())
```

```python
a=np.array([18,19,16,20,45,28]).reshape(3,2)
print(a)
print(a.max(axis=0))#column wise bigger number
print(a.max(axis=1))#row wise bigger number
```

```python
a=np.array([18,19,16,20,45,28]).reshape(3,2)
print(a)
print(a.min(axis=0))#column wise lower number
print(a.min(axis=1))#row wise lower number
```

```python
#joining arrays vertically
a=np.array([15,9,20,19]).reshape(2,2)
b=np.array([3,8,19,29]).reshape(2,2)
print(a)
print(b)
print(np.vstack((a,b)))
```

```python
#joining arrays horizontally
a=np.array([15,9,20,19]).reshape(2,2)
b=np.array([3,8,19,29]).reshape(2,2)
print(a)
print(b)
print(np.hstack((a,b)))
```

```python
a=np.array([15,9,20,19]).reshape(2,2)
b=np.array([3,8,19,29]).reshape(2,2)
print(a)
print(b)
print(np.stack((a,b),axis=0))
```

```python
a=np.array([15,9,20,19]).reshape(2,2)
b=np.array([3,8,19,29]).reshape(2,2)
print(a)
print(b)
print(np.stack((a,b),axis=1))
```

SEARCHING

```python
A=np.array([32,2,4,5,21,8])
print(np.where(A==8))
```

```python
A=np.array([32,2,4,5,21,8])
print(np.where(A>8))
```

```python
A=np.array([32,2,4,5,21,8])
print(np.searchsorted(A,4))
```

```python
A=np.array([5,8,74,14])
B=np.array([50,7,88,1])
print("Sum: ",np.add(A,B))
print("Difference:",np.subtract(A,B))
print("Product:",np.multiply(A,B))
print("Quotient:",np.divide(A,B))
print("Remainder:",np.mod(A,B))
print(np.divmod(A,B))
```

```python
A=np.array([5,8,74,14])
print(np.sort(A))
```

SET FUNCTIONS


```python
A=np.array([5,8,74,14])
B=np.array([50,7,88,1])
print(np.union1d(A,B))
```

```python
A=np.array([5,8,74,14])
B=np.array([50,7,88,1])
print(np.setdiff1d(A,B))
```

```python
A=np.array([5,8,74,14])
B=np.array([50,8,88,1])
print(np.intersect1d(A,B))
```

MORE FUNCTIONS

```python
#Comulative Sum
a=np.array([9,3,4,12,5,3])
print(np.cumsum(a))
```

```python
#Comulative Product
a=np.array([9,3,4,12,5,3])
print(np.cumprod(a))
```

```python
#LCM & GCD
print(np.lcm(25,50))
print(np.gcd(35,90))
```

```python
#LCM & GCD for mutiple elements
a=np.array([9,3,4,12,5,3])
print("LCM:",np.lcm.reduce(a))
print("GCD:",np.gcd.reduce(a))
```

```python
a=np.array([[1,2],[3,4]])
print(np.linalg.inv(a))
```
