EXTRACTING HTML CODE OF A WEBPAGE
- Import libraries
- Save the url
- Using the requers.get(),access the webpage
- using BeautifulSoup(),access the HTML code

```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as re
```

```python
url="https://www.snapchat.com/"
```

```python
page=re.get(url)
```

```python
code=bs(page.text,"html")
print(code)
```

```python
t=code.find("title")
print(t)
```

```python
print(t.text)
```

```python
u="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page=re.get(u)
c=bs(page.text,"html")
print(c)
```

```python
t=c.find("table")
print(t)
```

```python
h=t.find_all("th")
print(h)
```

```python
Headings=[i.text for i in h]
print(Headings)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=Headings)
print(mydf)
```

```python
rows=t.find_all("tr")
print(rows)
```

```python
td=t.find_all("td")
print(td)
```

```python
td_act=[i.text for i in td]
```

```python
print(td_act)
```

```python
ind=0
st=0
stop=4
for i in td_act:
  if stop <=40:
    mydf.loc[ind]=td_act[st:stop]
    st=st+4
    stop=stop+4
    ind=ind+1
print(mydf)
```

```python
mydf.to_csv("rich.csv")
```
